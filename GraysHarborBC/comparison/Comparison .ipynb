## Compare 3 methods, Randall (ocean forcing), Simple and Riemann.

Move 4 files simple1102, simple1187, riemann1102 and riemann1187 generated by SimpleKingTideBoundary.ipynb and RiemannKingTideBoundary.ipynb to this directory and run following code

```python
time = np.genfromtxt ('hours', delimiter=",")

randall1102 = np.genfromtxt ('randall1102', delimiter=",")
randall1187 = np.genfromtxt ('randall1187', delimiter=",")

simple1102 =np.genfromtxt ('simple1102', delimiter=",")
simple1187 = np.genfromtxt ('simple1187', delimiter=",")

riemann1102 =np.genfromtxt ('riemann1102', delimiter=",")
riemann1187 =np.genfromtxt ('riemann1187', delimiter=",")
```

```python
figure(figsize=(13,5))

plot(time,randall1102, label='Randall1102')  
plot(time,simple1102, label='simple1102')  
plot(time,riemann1102, label='Riemann1102')    


xlim(0,72)
ylim(-0.5,0.5)
legend(loc='upper left', fontsize=8)
xlabel('hours')
ylabel('m')
title('NOAA minus GeoClaw at 1102, Dec. 21-23, 2015')
grid(True)
```

```python
figure(figsize=(13,5))

plot(time,randall1187, label='Randall1187')  
plot(time,simple1187, label='simple1187')  
plot(time,riemann1187, label='Riemann1187')    



xlim(0,72)
ylim(-0.5,0.5)
legend(loc='upper left', fontsize=8)
xlabel('hours')
ylabel('m')
title('NOAA minus GeoClaw at 1187, Dec. 21-23, 2015')
grid(True)

```

```python
print("Randall, 1 norm error at 1102:", numpy.linalg.norm(randall1102,1))
print("Simple, 1 norm error at 1102:",numpy.linalg.norm(simple1102,1))
print("Riemann, 1 norm error at 1102:",numpy.linalg.norm(riemann1102,1))

print("Randall, 2 norm error at 1102:", numpy.linalg.norm(randall1102,2))
print("Simple, 2 norm error at 1102:",numpy.linalg.norm(simple1102,2))
print("Riemann, 2 norm error at 1102:",numpy.linalg.norm(riemann1102,2))
```

## Analysis
The simple method should work about the same as the ocean forcing method in terms of accuracy, but it takes less computational power since there is no source term involved.

The Riemann method doesn't work very well. Excluding the tidal signal, the water height decreases over time at 1102 and 1187. This might be caused by setting $q_l = q_m$ in our Riemann problem, further investigation is needed.
